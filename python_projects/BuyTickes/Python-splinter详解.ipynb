## 1. Splinter的简介 

   Splinter是一个使用Python开发的开源Web应用测试工具，它可以帮你实现自动浏览站点和与其进行交互。
   splinter.brower是一个开源工具，通过Python自动化测试ｗｅｂ，通过电脑自动操作网页。
## 2. Splinter的安装

  在安装好`python` 环境，并安装了pip

  直接执行以下命令：

  `pip install splinter`

  这是简单的安装，如果想用安装包安装，请自行搜索网上。

## 3. 检验Splinter是否安装成功

在`python`模式下输入命令：

`from splinter.browser import Browser`

如果没有报错，则说明安装成功！


## 4. 浏览器驱动的安装 
使用不同的浏览器要安装不同的驱动

Splinter默认支持的是Firefox的浏览器，因此如果你没有安装对相应版本的Firefox浏览器，就会报错。

## 5. Splinter的基本功能及操作示例

 splinter.brower基础知识

一、定位元素：
    - 唯一性定位：id
    - 控件名定位：name
    - 样式型定位：class
    - 值定位：value
    - 内容型定位：text
    - 超链接定位：href
    - 路径式定位：xpath
    - 其他
    
二、常用的API：
    
```
    browser.find_by_id('id')
 
    browser.find_by_name('name')
    
    browser.find_by_css('.h1')
    
    browser.find_by_tag('tag_name')
    
    browser.find_by_value('value')
    
    browser.find_by_xpath('path_name')
    
    
   
    #寻找链接如果您需要查找页面中的链接，可以使用
    
    find_link_by_text，
    find_link_by_partial_text，
    find_link_by_href或 find_link_by_partial_href方法。
  
```  
例子：
    
 ```
    links_found = browser.find_link_by_text('Link for Example.com')

    links_found = browser.find_link_by_partial_text('for Example')

    links_found = browser.find_link_by_href('http://example.com')

    links_found = browser.find_link_by_partial_href('example')    
```
三、操作元素
    
    常规HTML控件：
        
    - text单行文本框，textarea多行文本框
    - select下拉框：选择事件
    - redio单选框：单选事件
    - CheckBox复选框：复选事件
    - button按钮：点击事件
    - a超链接：跳转事件
    
    对应的API:
    
    ```
        click(点击事件)
        
        fill(填充事件)
        
        type(键入式填充事件，类似鼠标上下)
        
        select(下拉框选择事件)
        
        choose(单选框单选事件)
        
        check(复选框复选事件)
        
    ```

四、Cookies管理
    
    ```
        #Cookies添加：
        browser = Browser('chrome')
        
        browser.cookies.add({'username':"sd"})
    
        #cookies删除 delete(*cookies)
        
        browser.cookies.delete('username')
    ```






五、操作示例：

 ```
    #1.导入Browser模块
    from splinter import Browser

    #2.初始化browser,默认使用Firefox浏览器

    browser = Browser('firefox')

    #2.1从Firefox55开始，我们可以在Linux中运行Firefox作为无头浏览器。

    browser = Browser('firefox',headless=True)

    #2.3您可以使用profile关键字（将配置文件的名称作为str实例传递）指定在浏览器功能上使用的

    browser = Browser('firefox',profile='my_profile')

    #如果您不指定配置文件，则会创建一个新的临时配置文件（并在关闭浏览器时将其删除）。

    #使用Browser.visit进行浏览

    browser.visit('http://www.baidu.com')

    #输入搜索文本
    browser.fill('wd','python')

    #按搜索按钮
    browser.find_by_id('su').click()

    #一个CookieManager就像一个字典，所以你可以通过[]运算符来访问cookie的值，传递cookie标识符。


```
